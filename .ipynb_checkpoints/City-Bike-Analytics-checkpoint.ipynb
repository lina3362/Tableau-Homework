```python
clean_data_2019 = pd.concat([pd.read_csv(f) for f in glob.glob('data/2019/*.csv')], ignore_index = True)
clean_data_2020 = pd.concat([pd.read_csv(f) for f in glob.glob('data/2020/*.csv')], ignore_index = True)
```
